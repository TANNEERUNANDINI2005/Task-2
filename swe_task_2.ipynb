```python

with open("single_input.txt", "w") as f:
    f.write("50")


try:
    with open("single_input.txt", "r") as f:
        num = int(f.read().strip())
        print("Number read:", num)
        print("Double =", num * 2)
        print("Square =", num ** 2)
except FileNotFoundError:
    print("File not found!")
except ValueError:
    print("Error converting file content to integer.")
```

```python

with open("multiple_inputs.txt", "w") as f:
    f.write("4\n8\n12\n16")

try:
    with open("multiple_inputs.txt", "r") as f:
        numbers = [int(line.strip()) for line in f.readlines()]
        print("Numbers read:", numbers)
        print("Sum =", sum(numbers))
        print("Average =", sum(numbers)/len(numbers))
        print("Maximum =", max(numbers))
        print("Minimum =", min(numbers))
except FileNotFoundError:
    print("File not found!")
except ValueError:
    print("Error converting file content to integers.")
```

```python

num = 5
fact = 1
for i in range(1, num+1):
    fact *= i
print(f"Factorial of {num} = {fact}")

try:
    n = int(input("Enter a number for factorial: "))
    fact = 1
    for i in range(1, n+1):
        fact *= i
    print(f"Factorial of {n} = {fact}")
except ValueError:
    print("Invalid input! Please enter an integer.")
```

```python

word = "madam"
if word == word[::-1]:
    print(f"{word} is a palindrome")
else:
    print(f"{word} is not a palindrome")

user_word = input("Enter a word: ")
if user_word == user_word[::-1]:
    print(f"{user_word} is a palindrome")
else:
    print(f"{user_word} is not a palindrome")
```

```python

start, end = 1, 50
primes = []
for num in range(start, end+1):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            primes.append(num)
print("Prime numbers between 1 and 50:", primes)
```

```python

def calculator():
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")

    try:
        choice = int(input("Enter choice (1-4): "))
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))

        if choice == 1:
            print("Result =", a + b)
        elif choice == 2:
            print("Result =", a - b)
        elif choice == 3:
            print("Result =", a * b)
        elif choice == 4:
            if b != 0:
                print("Result =", a / b)
            else:
                print("Error: Division by zero!")
        else:
            print("Invalid choice!")
    except ValueError:
        print("Invalid input! Please enter integers.")

calculator()
```

```python



numbers = [45, 12, 78, 34, 23, 89, 5]

print("Original List:", numbers)

asc_sorted = sorted(numbers)
print("Ascending Order:", asc_sorted)


desc_sorted = sorted(numbers, reverse=True)
print("Descending Order:", desc_sorted)


try:
    user_input = input("Enter numbers separated by spaces: ")
    user_numbers = [int(num) for num in user_input.split()]
    print("User List:", user_numbers)
    print("Ascending:", sorted(user_numbers))
    print("Descending:", sorted(user_numbers, reverse=True))
except ValueError:
    print("Invalid input! Please enter integers only.")
```
